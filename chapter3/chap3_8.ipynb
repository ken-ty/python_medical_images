## 3.8 ピクセル（ボクセル）のキャリブレーション(校正)


ピクセルスペーシングのキャリブレーションをすることで、大きさの推定が行える。

手順:
- 大きさが既知のもの(コイン)のピクセル数を求める。
- 調べたい対象のピクセル数を求める。
- 基準の長さ/基準のpixel数 を求める。
- 対象の長さを求める。

リスト3.8.1 オヒルギ葉の直径と短径を求める


```
# オヒルギの長径と短径を求める
import plotly.express as px
import cv2 
# 表示画像上にカードるを当て、ピクセル位置を確認する
bgr = cv2.imread('/content/drive/My Drive/DATASET/dataset/Ohirugi-1.png',
                 cv2.IMREAD_UNCHANGED)
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
fig = px.imshow(rgb)
fig.show()
```

```
# 100円玉の直径 [380, 165], [380, 250]
# オヒルギ葉の長径 [262, 342], [172,70]
# オヒルギ葉の短径[138, 212], [263, 178]
```

```
# 続き
# コインの上端は(380, 165), 下端は(380, 250)であった。
import math
c_p_1 = (380, 165)
c_p_2 = (380, 250)
coin_len = math.sqrt( (c_p_1[0] - c_p_2[0])**2 + (c_p_1[1] - c_p_2[1])**2)
print("100円玉の直径ピクセル数", coin_len)

# 長径
o_long_p_1 = (262, 342)
o_long_p_2 = (172, 70)
longAxesPix = math.sqrt( (o_long_p_1[0] - o_long_p_2[0])**2 + (o_long_p_1[1] - o_long_p_2[1])**2)
print("長軸ピクセル数", longAxesPix)

# 短径
o_short_p_1 = (138, 212)
o_short_p_2 = (263, 178)
shortAxesPix = math.sqrt( (o_short_p_1[0] - o_short_p_2[0])**2 + (o_short_p_1[1] - o_short_p_2[1])**2)
print("長軸ピクセル数", shortAxesPix)

# 1ピクセル当たりの距離
COIN_LEN_100 = 22.6
pixspace = COIN_LEN_100 / coin_len # [mm]
print("１ピクセル当たりの距離は","{:.2}".format(pixspace), "mmです")

# オヒルギの長径と短径
print("長径 [cm]", longAxesPix*pixspace/10,"\n" 
      "短径 [cm]", shortAxesPix*pixspace/10)
```
